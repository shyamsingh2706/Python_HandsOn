```python
import pandas as pd
```

#### Read all the data w.r.t. Roster , Exam Grades and Quiz_grades

```python
roster_df = pd.read_csv("./Downloads/pandas-gradebook-project/data/Roster/roster.csv")
Exam_Grades_df = pd.read_csv("./Downloads/pandas-gradebook-project/data/Exam_Grades/hw_exam_grades.csv")
```

```python
roster_df
```

```python
Exam_Grades_df
```

```python
import os
import numpy as np

grades_list = []

for j in os.listdir("./Downloads/pandas-gradebook-project/data/Quiz_Grades/"):
      path = "./Downloads/pandas-gradebook-project/data/Quiz_Grades/" + j
      suffix = 'Grade_' + j.split('_')[1] 
      df  = pd.read_csv(path, index_col=None).rename(columns={"Grade": suffix}) 
      grades_list.append(df) 
    
Quiz_grades_df = pd.concat(grades_list)
Quiz_grades_df
```

#### Apply Data Cleansing to bring all Quiz Grades for one student into single Row from 5 different rows

```python
Quiz_grades_df = Quiz_grades_df.sort_values(by=['First Name','Last Name','Email'] ).replace(np.NaN,0)  
Quiz_grades_df = Quiz_grades_df.groupby(['First Name','Last Name','Email']).max()
Quiz_grades_df = Quiz_grades_df.reset_index()
Quiz_grades_df 
```

#### Convert all the attributes that will be part of join into common format across DFs

```python
roster_df['Email Address'] = roster_df['Email Address'].str.lower()
roster_df['NetID'] = roster_df['NetID'].str.lower()
```

#### Reset Index for all DFs in order to merge the DFs and avoid duplicate attributes after merging of DFs

```python
Exam_Grades_df.set_index('SID',inplace = True) 
roster_df.set_index('NetID',inplace=True)
Quiz_grades_df.set_index('Email',inplace=True)
```

#### Merge Roster DF & Exam_grades_df using SID

```python
Student_exam_grades_df = pd.merge(left=roster_df, right=Exam_Grades_df, left_index=True, right_index=True)
Student_exam_grades_df.head(2)
```

#### Merge Student_exam_grades_df with Quiz_grades_df on Email Address to create final DF that contins all Data for all students

```python
Student_all_grades_df = pd.merge(left=Student_exam_grades_df, right=Quiz_grades_df, left_on='Email Address', right_index=True)
Student_all_grades_df.head(2)
```

#### Find out NaN values present in Final DF and replace that will 0

```python
Student_all_grades_df.isna().sum()
```

```python
Student_all_grades_df = Student_all_grades_df.fillna(0)
```

### Calculate Grades

#### Calculate Exam Scores 

```python
n_exams = 3
for n in range(1, n_exams + 1):
    Student_all_grades_df[f"Exam {n} Score"] = (
        Student_all_grades_df[f"Exam {n}"] / Student_all_grades_df[f"Exam {n} - Max Points"]
    )
```

```python
pd.set_option("display.max_columns" , 65)
Student_all_grades_df.head(2)
```

#### Calculating the Homework Scores

#### By total score

```python
Student_all_grades_df['total Homework Score'] = Student_all_grades_df.filter(regex='^Homework\s\d+$').sum(axis=1)
Student_all_grades_df['total Homework Max Points'] = Student_all_grades_df.filter(regex='^Homework.*Points$').sum(axis=1)
Student_all_grades_df["Total Homework"] = Student_all_grades_df['total Homework Score'] / Student_all_grades_df['total Homework Max Points']
```

```python
Student_all_grades_df.head(2)
```

#### By average score

```python
n_homework = 10
for n in range(1, n_homework + 1):
    Student_all_grades_df[f"Homework {n} Avg"] = (
        Student_all_grades_df[f"Homework {n}"] / Student_all_grades_df[f"Homework {n} - Max Points"]
    )

Student_all_grades_df['Average Homework'] = Student_all_grades_df.filter(regex='^Homework.*Avg').sum(axis=1) / n_homework
```

#### Pick the max of Total Homework Score vs Avg Homework Score and Assign to Final Homework Score for each student

```python
pd.set_option("display.max_columns" , 80)
Student_all_grades_df["Homework Score"] = Student_all_grades_df[
    ["Total Homework", "Average Homework"]
].max(axis=1)
Student_all_grades_df.head(2)
```

#### Calculating the Quiz Score 

#### By Total Score

```python
quiz_max_points = pd.Series(
    {"Quiz 1": 11, "Quiz 2": 15, "Quiz 3": 17, "Quiz 4": 14, "Quiz 5": 12}
)

Student_all_grades_df['total Quiz Score'] = Student_all_grades_df.filter(regex='^Grade\_\d+$').sum(axis=1) 
Student_all_grades_df['total Quiz Max Points']    = quiz_max_points.sum() 
Student_all_grades_df['total Quiz']    = Student_all_grades_df['total Quiz Score'] / Student_all_grades_df['total Quiz Max Points'] 
```

```python
Student_all_grades_df.head(2)
```

#### By Avg score

```python
n_quiz = 5
for n in range(1, n_quiz + 1):
    Student_all_grades_df[f"Quiz {n} Avg"] = (
        Student_all_grades_df[f"Grade_{n}"] / quiz_max_points[f"Quiz {n}"]
    )

Student_all_grades_df['Average Quiz'] = Student_all_grades_df.filter(regex='^Quiz.*Avg').sum(axis=1) / n_quiz
```

```python
pd.set_option("display.max_columns" , 80)
Student_all_grades_df["Quiz Score"] = Student_all_grades_df[["total Quiz", "Average Quiz"]].max(axis=1)
```

```python
Student_Final_Score_df = Student_all_grades_df[['First Name_x','Last Name_x','Name','Section','Email Address', 'Exam 1 Score','Exam 2 Score','Exam 3 Score','Homework Score','Quiz Score']]
Student_Final_Score_df
```

#### Calculate the grades

```python
weightings = pd.Series(
    {
        "Exam 1 Score": 0.05,
        "Exam 2 Score": 0.1,
        "Exam 3 Score": 0.15,
        "Homework Score": 0.4,
        "Quiz Score": 0.30
    }
)
weightings
```

#### Approach 1 to calculate final score by hardcoding Column Name

```python
Student_Final_Score_df['Final_Score'] =  (Student_Final_Score_df['Exam 1 Score'] * weightings['Exam 1 Score']) +  (Student_Final_Score_df['Exam 2 Score'] * weightings['Exam 2 Score']) +  (Student_Final_Score_df['Exam 3 Score'] * weightings['Exam 3 Score']) +  (Student_Final_Score_df['Homework Score'] * weightings['Homework Score'])+  (Student_Final_Score_df['Quiz Score'] * weightings['Quiz Score'])
Student_Final_Score_df
```

#### Approach 2 to calculate final score by using weightings Index  Method

```python
Student_Final_Score_df["Final Score"] = (Student_Final_Score_df[weightings.index] * weightings).sum(
    axis=1
)
Student_Final_Score_df
```

```python
import numpy as np

Student_Final_Score_df["Ceiling Score"] = np.ceil(Student_Final_Score_df["Final Score"] * 100)
Student_Final_Score_df.head(2)
```

#### Assign Grades

```python
def Assign_grades(score):
    
    Grade = ''
    
    if (( score < 60 )):
        Grade = 'F'
    elif ((score >= 60 ) and (score < 70 )):
        Grade = 'D'
    elif ((score >= 70 ) and (score < 80 )):
        Grade = 'C'
    elif ((score >= 80 ) and (score < 90 )):
        Grade = 'B'
    elif ((score >= 90 )):
        Grade = 'A'
    
    return Grade
```

```python
Student_Final_Score_df["Grade"] = Student_Final_Score_df["Ceiling Score"].apply(Assign_grades) 
```

```python
Student_Final_Score_df = Student_Final_Score_df.rename(columns = {'First Name_x': 'First Name', 'Last Name_x':'Last Name'})
Student_Final_Score_df.head(10)
```

#### separate the students into each section and sort them by their last name

```python
section_df = Student_Final_Score_df.sort_values(by='Last Name').groupby('Section') 
```

```python
for section in section_df.groups :
    df = section_df.get_group(section)
    section_file = f"Section {section} Grades.csv"
    df.to_csv(section_file,index=True,index_label ='SID')
```
